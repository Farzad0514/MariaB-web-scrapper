```python
l=[]
base_url=("https://www.mariab.pk/ready-pret/casuals.html#/page/")
for page in range(1,10):
    print(base_url+str(page))
    site=requests.get(base_url+str(page))
    #print(site.status_code)
    soup=BeautifulSoup(site.content, "html.parser")
    #print(soup.prettify())
    all=soup.find_all("div", {"class": ["product-info"]})
    #len(all)

    for item in all: 
        d={}
        #name
        d["item"]= item.find("h3", {"class", "product-name"}).text 

        #old-price
        try:  d["oldprice"]= item.find_all("span", {"class", "price"})[0].text.replace(" ", "").replace("PKR", "").replace("\n", "")
        except: d["oldprice"]= None
        #special-price
        try: d["special-price"]=item.find_all("span", {"class", "price"})[1].text.replace(" ", "").replace("PKR", "").replace("\n", "")
        except:d["special-price"]= None

        #size
        sizedata= item.find("div", {"class", "available-sizes"}) 
        size=sizedata.find_all("li") 
        d["size01"]=size[0].text
        try: d["size02"]=size[1].text 
        except:d["size02"]= None 
        try: d["size03"]=size[2].text
        except: d["size03"]=None 
        try: d["size04"]=size[3].text
        except: d["size04"]=None 
        try: d["size05"]=size[4].text 
        except: d["size05"]=None 
        try: d["size06"]=size[5].text 
        except: d["size05"]= None 

        #url
        d["url"]=item.find("h3", {"class", "product-name"}).a.get('href')

        #print()
        l.append(d)
    
```
